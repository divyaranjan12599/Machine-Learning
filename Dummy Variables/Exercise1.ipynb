```python
import pandas as pd
df = pd.read_csv('exercise.csv')
df
```

```python
import matplotlib.pyplot as plt
plt.scatter(df.Mileage, df['Sell Price($)'])
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(df[['Mileage']], df['Sell Price($)'])
```

```python
plt.scatter(df.Mileage, df['Sell Price($)'])
plt.plot(df['Mileage'], model.predict(df[['Mileage']]), color = '#000088')
```

```python
dummy_models = pd.get_dummies(df['Car Model'])
```

```python
dummy_models
```

```python
merged = pd.concat([df,dummy_models],axis='columns')
```

```python
merged
```

```python
final = merged.drop(['Car Model', 'Mercedez Benz C class'], axis = 'columns')
final
```

```python
X = final.drop(['Sell Price($)'], axis = 'columns')
X
```

```python
y = final['Sell Price($)']
```

```python
model.fit(X,y)
```

```python
model.score(X,y)
```
